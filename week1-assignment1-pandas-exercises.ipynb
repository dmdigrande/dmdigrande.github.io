##### 4. Calculate the totals of the `freight`, `mail`, and `passengers` columns for flights from the United Kingdom to the United States.

```python
uk_to_us_flights = flights[(flights['ORIGIN_COUNTRY_NAME'] == 'United Kingdom') & (flights['DEST_COUNTRY_NAME'] == 'United States')]
uk_to_us_flights
```

```python
total_freight = uk_to_us_flights['FREIGHT'].sum()
total_mail = uk_to_us_flights['MAIL'].sum()
total_passengers = uk_to_us_flights['PASSENGERS'].sum()

print('TOTAL FREIGHT:' ,  total_freight)
print('TOTAL MAIL:' , total_mail)
print('TOTAL PASSENGERS:' , total_passengers)
```

##### 5. Find the top 5 carriers for median route distance (i.e., for all origin-destination pairs that each carrier has, find the median distance after removing duplicates).

```python
unique_carrier = flights.drop_duplicates(subset=['CARRIER','ORIGIN', 'DEST'])
```
