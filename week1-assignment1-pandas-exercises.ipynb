##### 4. Calculate the totals of the `freight`, `mail`, and `passengers` columns for flights from the United Kingdom to the United States.

```python
uk_to_us_flights = flights[(flights['ORIGIN_COUNTRY_NAME'] == 'United Kingdom') & (flights['DEST_COUNTRY_NAME'] == 'United States')]
uk_to_us_flights
```

```python
total_freight = uk_to_us_flights['FREIGHT'].sum()
total_mail = uk_to_us_flights['MAIL'].sum()
total_passengers = uk_to_us_flights['PASSENGERS'].sum()

print(total_freight)
print(total_mail)
print(total_passengers)
```

##### 5. Find the top 5 carriers for median route distance (i.e., for all origin-destination pairs that each carrier has, find the median distance after removing duplicates).

```python
unique_carrier = flights.drop_duplicates(subset=['CARRIER','ORIGIN', 'DEST'])

unique_carrier.head()
```

```python
unique_carrier['MEDIAN_DISTANCE'] = unique_carrier['DISTANCE'].median()

median_distance_df = unique_carrier.sort_values(by='MEDIAN_DISTANCE', ascending=False)

median_distance_df.head()
```

##### 6. Find the total cargo transported (mail + freight) and mean distance traveled for the 10 carriers that transported the most cargo.

```python
unique_carrier = unique_carrier.assign(
    TOTAL_CARGO = lambda x: x.MAIL + x.FREIGHT,
    MEAN_DISTANCE = lambda x: x.DISTANCE.mean())

total_cargo_sort = unique_carrier.sort_values(by='TOTAL_CARGO', ascending=False)

total_cargo_sort.head(10)
```

##### 7. Which 10 carriers flew the most passengers out of the United States to another country?

```python
unique_c= flights.drop_duplicates(subset=['CARRIER'])

us_flights = unique_c[(unique_c['ORIGIN_COUNTRY_NAME'] == 'United States')]

us_flights_sorted = us_flights.sort_values(by='PASSENGERS', ascending=False)

us_flights_sorted.head(10)
```

8. For each of the carriers found in *#7*, find the most popular destination country outside of the United States.

```python
#source: https://stackoverflow.com/questions/39922986/how-do-i-pandas-group-by-to-get-sum

popular_destination = us_flights[us_flights['DEST_COUNTRY_NAME'] != 'United States']

popular_destination = popular_destination.groupby(['DEST_CITY_NAME','PASSENGERS']).sum()

popular_destination = popular_destination.sort_values(by='PASSENGERS', ascending=False)
popular_destination.head(1)
```
